# Baixando contratos da Prefeitura de São Paulo de 2014

Desde 2014 a Prefeitura de São Paulo disponibiliza os contratos integrais com empresas, convênios com organizações, locação de imóveis, entre outros tipos no site :

 * http://transparencia.prefeitura.sp.gov.br/contas/Paginas/Contratos-v2.aspx
 
O problema é que navegando pelo sistema facilmente dá para notar o quão lento é realizar cada ação neste site como simplesmente mudar de página. Além disso não é possível buscar os contratos por um período, ver informações sobre os tipos de arquivos ou uma discriminação mais detalhada sobre o que se refere o contrato sem necessitar ter que ler o contrato inteiro.

Seria muito interessante por exemplo se fosse possível ter associado a cada contrato dados como tipo do objeto, quantidade, se fosse uma locação, construção ou manutenção de área pública, endereço e coordenadas geográficas para serem utilizadas em sistemas como o OpenStreeMaps.

Como não possuímos isso, achei que seria interessantes baixar todos os contratos para depois pensar em métodos para extrair informações relevantes deles. Quem sabe iniciar um projeto de crowdsourcing usando PyBossa, ou tentar algumas técnicas de extração automática de informações dos documentos.

Para baixar os contratos pensei em dois possíveis caminhos:

 1. Criar um programa que utilize a biblioteca [Selenium](http://www.seleniumhq.org/) que permite automatizer um webbrowser, fazendo com que tal programa fosse navegando pelo site da prefeitura e clicando em todos os links para baixar os contrato.
 
 2. Criar um programa que não utilizasse um webbrowser mas realizasse as requisições diretamente para o site da prefeitura.
 
Optei pela segunda opção, mas confesso que talvez a opção 1 teria sido mais simples por alguns probleminhas que surgiram no meio do caminho.

Mas vamos lá, uma breve descrição de como os dados foram baixados.

# Fazendo as requisições sem utilizar um web browser

A primeira coisa a ser feita foi utilizar o Firefox Developer Tools para analisar como as requisições estavam sendo feitas para iniciar o download de um contrato. Infelizmente, cada contrato não possui um URL único, pois neste caso bastaria obter essa lista de URL's para baixar usando um programa como curl ou wget. Pra piorar, as requisições de download eram links que executavam um código em Javscript, links como: 

    javascript:__doPostBack('ctl00$ContentPlaceHolder1$grdContratos$ctl02$lnkIntegra','') 
    
A informação do número do contrato encontra-se no número que segue o segundo 'ctl' no texto acima. Neste exemplo, o contrato referenciado é o 'ctl02', que é sempre o primeiro contrato numa das páginas de contratos da prefeitura.

Utilizei a ferramenta Network do Developer Tools do Firefox para analisar as requisições HTTP realizadas ao pedir o download de um contrato. Com isso descobri que o pedido de download era realizado ao endereço:

 http://transparenciasp.prefeitura.sp.gov.br/sj2224_contrato/PaginasPublicas/frmPesquisaContrato.aspx
 
Mas para entender como a informação de qual contrato está sendo pedido ao servidor, analisei os dados em Params dessa requisição ainda dentro da ferramenta Network.

```python
contratos_types = {}

for c in contratos_files:
    cf = !file $c
    cdata = cf[0].partition(':')
    contratos_types[cdata[0]] = cdata[2].strip()
```

```python
# Algumas conversões úteis para os nomes, para não termos contagem em grupos diferentes para PDF 1.5, PDF 1.6, etc

def conv_name(name):
    if 'Word' in name or 'Composite Doc' in name:
        return 'Word'
    elif 'PDF' in name:
        return 'PDF'
    elif 'ASCII' in name or 'ISO' in name:
        return 'TXT'
    elif 'TIFF' in name:
        return 'TIFF'
    elif 'HTML' in name:
        return 'HTML'
    elif 'JPEG' in name:
        return 'JPEG'
    elif 'Rich' in name:
        return 'RTF'
    elif 'Zip' in name:
        return 'ZIP'
    elif 'Open' in name:
        return 'ODF'
    else:
        return name
```

```python
ctypes_hist = Counter([conv_name(c) for c in contracts_types.values()])
```
