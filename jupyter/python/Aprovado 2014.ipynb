```python
%pylab inline

import pandas as pd
import seaborn as sb
```

```python
data = pd.read_excel('../Orcamento/2014/aprovado2014.xls')
```

```python
orgaos = data['Ds_Orgao'].value_counts()
```

```python
funcoes = data['Ds_Funcao'].value_counts()
```

# Valor por Função (2014)

```python
s = data.groupby('Ds_Funcao')['Sld_Orcado_Ano'].sum()
s.sort()
s.plot(kind='barh')
```

# Valor por Fonte (2014)

```python
dfonte = data.groupby('Ds_Fonte')['Sld_Orcado_Ano'].sum()
dfonte.sort()
dfonte.plot(kind='barh')
```

# Agrupando por Órgãos

```python
dorgao = data.groupby('Ds_Orgao')['Sld_Orcado_Ano'].sum()
dorgao.sort()
```

## Fundos

```python
dorgao.filter(regex='^Fundo').plot(kind='barh')
```

## Secretarias

```python
dorgao.filter(regex='^Secretaria').plot(kind='barh')
```

## Subprefeituras

```python
dorgao.filter(regex='^Sub').plot(kind='barh')
```

## Secretaria x Subprefeituras

```python
sub_total = dorgao.filter(regex='^Sub').sum()
sec_total = dorgao.filter(regex='^Secretaria').sum()
total = float(sub_total + sec_total)
print "Subprefeituras: {:.2%}\nSecretarias: {:.2%}".format(sub_total/total, sec_total/total)
```

```python
print "Subprefeitura"
print "R$ {:,.2f}".format(sub_total)
```

```python
print "Secretaria"
print "R$ {:,.2f}".format(sec_total)
```
